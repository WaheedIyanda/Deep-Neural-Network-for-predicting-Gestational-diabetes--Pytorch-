# Data Preprocessing

```python
# since i have observed and imbalance in our data set, I use oversampling of the minority class
```

```python
from sklearn.model_selection import train_test_split
```

```python
y= gestational_data["Prediction"]
X= gestational_data.drop("Prediction", axis=1)
##############################
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42, stratify=y)
```

```python
##### using SMOTE
from imblearn.over_sampling import SMOTENC
##########
categorical_feature = [5]
```

```python
smote_nc = SMOTENC ( categorical_features=categorical_feature, random_state=42)
X_train_resampled, y_train_resampled = smote_nc.fit_resample(X_train, y_train)
print(pd.Series(y_train_resampled).value_counts())
```

# ML Training

Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train_resampled, y_train_resampled)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score, precision_score, recall_score
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

Building a Neural Network that get close to 95% Accuracy

```python
import torch
import torch.nn as nn 
from torch.utils.data import DataLoader, TensorDataset
import torch.optim as optim
```

```python
X_train_tensor = torch.tensor(X_train_resampled.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train_resampled.values, dtype=torch.float32).unsqueeze(1)
X_test_tensor = torch.tensor(X_test.values, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).unsqueeze(1)
```

```python
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
```

```python
hidden1_size = 16
hidden2_size = 32
hidden3_size = 16
hidden4_size =8
hidden5_size =4
output_size = 1
##############################################
class NeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden1_size, hidden2_size, hidden3_size, hidden4_size, hidden5_size, output_size):
        super(NeuralNetwork, self).__init__()
        self.layer1 = nn.Linear(input_size, hidden1_size)
        self.layer2 = nn.Linear(hidden1_size, hidden2_size)
        self.layer3 = nn.Linear(hidden2_size, hidden3_size)
        self.layer4 = nn.Linear(hidden3_size, hidden4_size)
        self.layer5 = nn.Linear(hidden4_size, hidden5_size)
        self.output = nn.Linear(hidden5_size, output_size)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.layer1(x))
        x = self.relu(self.layer2(x))
        x = self.relu(self.layer3(x))
        x = self.relu(self.layer4(x))
        x = self.relu(self.layer5(x))
        x = self.sigmoid(self.output(x))
        return x

inpute_size = X_train_resampled.shape[1]

my_model = NeuralNetwork(inpute_size, hidden1_size, hidden2_size, hidden3_size, hidden4_size, hidden5_size, output_size)
print(my_model)


```

```python
#####setting loss function and optimizer
criterion = nn.BCELoss()
optimizer = optim.Adam(my_model.parameters(), lr = 0.001)
```

```python
num_epochs = 80
loss_values = []

for epoch in range(num_epochs):
    my_model.train()
    epoch_loss = 0
    for batch_X, batch_y in train_loader:
        optimizer.zero_grad()
        outputs = my_model(batch_X)  # Make the prediction
        loss = criterion(outputs, batch_y)  # Compare the predicted value to the real value
        loss.backward()
        optimizer.step()
        
        epoch_loss += loss.item()
    
    avg_epoch_loss = epoch_loss / len(train_loader)
    loss_values.append(avg_epoch_loss)

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {avg_epoch_loss:.4f}')



plt.figure(figsize=(10, 5))
plt.plot(range(1, num_epochs + 1), loss_values, marker='o')
plt.title('Training Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

```python
my_model.eval()
with torch.no_grad():
    y_predi = my_model(X_test_tensor)
    y_pred_class = (y_predi > 0.5).float()

y_test_np = y_test_tensor.numpy()
y_pred_np = y_pred_class.numpy()

accuracy = accuracy_score(y_test_np, y_pred_np)
precision = precision_score(y_test_np, y_pred_np)
recall = recall_score(y_test_np, y_pred_np)
f1 = f1_score(y_test_np, y_pred_np)

print(f"Accuracy: {accuracy:.4f}")  
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```

```python
#saving the model
torch.save(my_model.state_dict(), 'my_model.pth')
print("Model trained and saved successfully!")
```
